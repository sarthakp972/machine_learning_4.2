```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/drive/MyDrive/Machine_learning_15 01 2024/Homeprices.csv')
```

```python
df.columns

```

#Linear Regression with multiple variable
 * y=m1x1 + m2x2 + m3x3 + c
 * price = m1arae + m2 room + m3 duration +c

 1. price= Dependent Variables
 2. area, room, duration = independent Variables
 3. m1,m2,m3 = cofficient
 4. c=intercept

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','room','duration ']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40 ]])
```

```python
sns.scatterplot(x='area', y='price', data=df, label='Area')
sns.scatterplot(x='room', y='price', data=df, label='Room')
sns.scatterplot(x='duration ', y='price', data=df, label='Duration')
plt.xlabel('Independent variables')
plt.legend()
plt.show()
```

```python
sns. set(style="whitegrid")

sns.lmplot(x='area',y='price', data=df, aspect=2, height=3)
plt.title('Linear Regression plot for Price vs. Area')
sns. set(style="whitegrid")

sns.lmplot(x='room',y='price', data=df, aspect=2, height=3)
plt.title('Linear Regression plot for room vs. price')

sns.lmplot(x='duration ',y='price', data=df, aspect=2, height=3)
plt.title('Linear Regression plot for Price vs. duration')

```
